with following customisations. Chart should have Month name on X-
axis instead of numbers.

Maker =”*”
Marker size=10
linestyle=”–“
Linewidth =3

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define month labels
month=['Jan','Feb','Mar','Apr','May','Jun']

# Prepare data
data = """2018,2019,2020,2021
2500,2600,2450,3250
2200,1800,2000,3200
3000,2500,3000,3800
2800,3000,3100,2800
2750,3200,1950,2900
3500,2000,3550,2300"""

with open('monthsales.csv', 'w') as m:
  m.write(data)

df = pd.read_csv('monthsales.csv')

plt.figure(figsize=(10,6))

for year in df.columns:
  plt.plot(month, df[year], marker='*',markersize=10,linewidth=3,label=year)

plt.title('Year-Wise-Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(title='Years')
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data="""State,Population,Lit_Rate,Region
India,1210854977,74.04,National
Andaman & Nicobar Islands,380581,86.63,South
Andhra Pradesh,84580777,67.02,South
Arunachal Pradesh,1383727,65.38,East
Assam,31205576,72.19,East
Bihar,104099452,61.80,East
Chandigarh,1055450,86.05,North
Chhattisgarh,25545198,70.28,East
Dadra & Nagar Haveli,343709,76.24,West
Daman & Diu,243247,87.10,West
Delhi,16787941,86.21,North
Goa,1458545,88.70,West
Gujarat,60439692,78.03,West
Haryana,25351462,75.55,North
Himachal Pradesh,6864602,82.80,North
Jammu & Kashmir,12541302,67.16,North
Jharkhand,32988134,66.41,East
Karnataka,61095297,75.36,South
Kerala,33406061,94.00,South
Lakshadweep,64473,91.85,South
Madhya Pradesh,72626809,69.32,North
Maharashtra,112374333,82.34,West
Manipur,2855794,76.94,East
Meghalaya,2966889,74.43,East
Mizoram,1097206,91.33,East
Nagaland,1978502,79.55,East
Odisha,41974218,72.87,East
Puducherry,1247953,85.85,South
Punjab,27743338,75.84,North
Rajasthan,68548437,66.11,West
Sikkim,610577,81.42,East
Tamil Nadu,72147030,80.09,South
Tripura,3673917,87.22,East
Uttar Pradesh,199812341,67.68,North
Uttarakhand,10086292,78.82,North
West Bengal,91276115,76.26,East"""

with open ('data.csv','w') as d:
  d.write(data)

df=pd.read_csv('data.csv')

#scatter plot
plt.figure(figsize=(12, 6))
regions = df['Region'].unique()

for region in regions:
    region_df = df[df['Region'] == region]

    sizes = np.sqrt(region_df['Lit_Rate']) * 10
    plt.scatter(region_df['Population'], region_df['Lit_Rate'],
                s=sizes, marker='D', label=region, alpha=0.7)

plt.title('Population vs Literacy Rate by Region')
plt.xlabel('Population')
plt.ylabel('Literacy Rate')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

#bar chart
plt.figure(figsize=(10, 6))
avg_lit = df.groupby('Region')['Lit_Rate'].mean()
avg_lit.plot(kind='bar', color='skyblue')

plt.title('Average Literacy Rate by Region')
plt.xlabel('Region')
plt.ylabel('Average Literacy Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

7. In order to assess the performance of students of a class in the annual
examination, the class teacher stored marks of the students in all the 5
subjects in a CSV “Marks.csv” file as shown in Table. Plot the data using
boxplot and perform a comparative analysis of performance in each
subject.

image.png

```python
import pandas as pd
import matplotlib.pyplot as plt

marks="""Name,English,Maths,Hindi,Science,Social_Studies
Rishika Batra,95,95,90,94,95
Waseem Ali,95,76,79,77,89
Kulpreet Singh,78,81,75,76,88
Annie Mathews,88,63,67,77,80"""

with open ('marks.csv','w') as m:
  m.write(marks)

df = pd.read_csv('marks.csv')

subject_data = df[['English', 'Maths', 'Hindi', 'Science', 'Social_Studies']]

plt.figure(figsize=(10, 6))
plt.boxplot(subject_data.values, labels=subject_data.columns, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'))

plt.title('Comparative Analysis of Performance in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks Obtained')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

8. Write a Python program to create bar plot of scores by group and gender.
Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
The code snippet gives the output shown in the following screenshot:

image.png

```python
import numpy as np
import matplotlib.pyplot as plt


groups = ['G1', 'G2', 'G3', 'G4', 'G5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]


x = np.arange(len(groups))
width = 0.4


fig, ax = plt.subplots(figsize=(8, 6))
bar1 = ax.bar(x - width/2, means_men, width, label='Men', color='green')
bar2 = ax.bar(x + width/2, means_women, width, label='Women', color='red')


ax.set_ylabel('Scores')
ax.set_xlabel('Person')
ax.set_title('Scores by Person')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()

ax.bar_label(bar1, padding=3)
ax.bar_label(bar2, padding=3)

fig.tight_layout()
plt.show()
```

10.Read the total profit of each month and show it using the histogram to

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('monthsales.csv')

profitList = df['total_profit'].tolist()
profit_range = [150000,175000,200000,225000,250000,300000,350000]

plt.figure(figsize=(8, 6))
plt.hist(profitList, profit_range, label='Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```
